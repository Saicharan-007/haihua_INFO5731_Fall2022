Question 2 (10 points): Write python code to collect 1000 data samples you discussed above.

```python
# You code here (Please add comments in the code):
import requests
from bs4 import BeautifulSoup
import pandas as pd


url_title = 'https://www.imdb.com/search/title/?title_type=feature' 
request = requests.get(url_title)
url_dt = BeautifulSoup(request.text, 'lxml') 
url_tg = url_dt.find_all('a', attrs={'class': None})
url_tg_dt = [tag.attrs['href'] for tag in url_tg 
              if tag.attrs['href'].startswith('/title') & tag.attrs['href'].endswith('/')]
url_tg_dt = list(dict.fromkeys(url_tg_dt)) 
url_lnk = ["https://www.imdb.com" + tag + 'reviews' for tag in url_tg_dt]
print(url_lnk)
print(url_lnk[0])
print(url_lnk[1])

def getindex(a):
    minposition = a.index(min(a))
    maxposition = a.index(max(a))
    return minposition,maxposition

comments = []
votes1 = []
vt1=[]
Uid=[]
cine=[]
review_nt=[]
for i in range(0,50):
    url_dt = BeautifulSoup(requests.get(url_lnk[i]).text, 'html.parser')
    voting = [tag.previous_element for tag in 
                           url_dt.find_all('span', attrs={'class': 'point-scale'})]
    if len(voting) > 0:   #If rating percentage is greater than 0
        n_index, p_index = getindex(list(map(int, voting)))
        review_lst = url_dt.find_all('a', attrs={'class':'title'})
        revw = ["https://www.imdb.com" + review['href'] for review in review_lst]
    else:
        print(None)



    for j in revw:
        txt=BeautifulSoup(requests.get(j).text, 'html.parser').find('div', attrs={'class': 'text show-more__control'}).get_text()
        rating=BeautifulSoup(requests.get(j).text, 'html.parser').find('div', attrs={'class': 'lister-item-content'}).get_text()
        movieusers=(BeautifulSoup(requests.get(j).text, 'html.parser').find('title').get_text())
        descr=(BeautifulSoup(requests.get(j).text, 'html.parser').find('a', attrs={'class': 'title'}).get_text())
        len_rating1=len(vt1)
        if(txt == ''):
            comments.append('No Review')
        else:
            comments.append(txt)
        votes1.append(rating)

        Uid.append(movieusers.split("'s")[0])
        cine.append(movieusers.split('Review of')[1])
        review_nt.append(descr)
  
    print('vote : '+str(len(votes1))+' review list:'+str(len(comments)))
    if(len_rating1==0):
        k=0
    else:
        k=len_rating1
    for j in range(k,len(votes1)):
        vt1.append(votes1[j].split('/10')[0].split('\n')[len(votes1[j].split('/10')[0].split('\n'))-1])
    print('rating1:'+str(len(vt1)))

data_review = {
        "Cinema name" : cine,
        "User Name" : Uid,
        "Review_Note" : review_nt,
        'Detail Reviews': comments,
        'Rating': vt1,
        
        }
df = pd.DataFrame(data_review)
df


```
